```python
data_training=pd.DataFrame(df.Close[0:int(len(df)*.7)])
print(data_training)
```

```python
data_testing=pd.DataFrame(df.Close[int(len(df)*.7):int(len(df))])
print(data_testing)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array=scaler.fit_transform(data_training)
```

```python
data_training_array
```

```python
x_train=[]
y_train=[]

for i in range(100,data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i,0])
```

```python
x_train,y_train=np.array(x_train),np.array(y_train)
```

```python
x_train.ndim
```

```python
y_train.ndim
```

```python
!pip install tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Input
```

```python
model=Sequential()
model.add(Input(shape=(100,1)))#1 is the feature(closing price) and 100 is the previous 100 days data!
```

```python
model.add(LSTM(units=128,activation='tanh',return_sequences=True))#important
model.add(LSTM(units=64))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train,y_train,epochs=50)
```

```python
model.summary()
```

```python
model.save("stock_prediction_model.keras")
```

```python
past_100_days=data_training.tail(100)
```

```python
past_100_days
```

```python
final_df=pd.concat([past_100_days,data_testing],ignore_index=True)
final_df
```

```python
input_data=scaler.fit_transform(final_df)
input_data
```

```python
x_test=[]
y_test=[]
for i in range(100,input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i,0])
    
```

```python
x_test,y_test=np.array(x_test),np.array(y_test)
```

```python
y_test
```

```python
y_predicted=model.predict(x_test)
```

```python
y_predicted
```

```python
y_predicted = scaler.inverse_transform(y_predicted.reshape(-1, 1)).flatten()
y_test = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
```

```python
y_predicted
```

```python
y_test
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test,'b',label='Original Price')
plt.plot(y_predicted,'r',label='Predicted Price')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_predicted)
```

```python
print(mse)
```

```python
rmse=np.sqrt(mse)
print(rmse)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_predicted)
print(r2)
```
